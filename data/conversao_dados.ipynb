## Conversão do banco de dados do INMETRO para json

```python
# Bibliotecas e arquivo Excel

import pandas as pd
import json

xlsx_path = "ar_condicionado_idrs.xlsx"
output_json = "equipamentos_integrados.json"

df = pd.read_excel(xlsx_path)
```

```python
# Remover NaN → None e converter para lista de dicionários

df = df.where(pd.notnull(df), None)

records = df.to_dict(orient="records")
```

```python
# Salvar JSON

with open(output_json, "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=3)

print("JSON gerado com sucesso em:", output_json)
```
